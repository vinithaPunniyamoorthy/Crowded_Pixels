```python
import cv2 
# Load an image 
img = cv2.imread('img_31.jpg') 
print('Datatype:', img.dtype, '\nDimensions:', img.shape)
# Display the image 
cv2.imshow("image", img) 
# Wait for the user to close the window 
cv2.waitKey(0) 
# Clean up 
cv2.destroyAllWindows()
import matplotlib.pyplot as plt
# Display the image 
plt.imshow(img)
plt.title('Displaying image using Matplotlib')
plt.show()
from cv2 import imwrite
imwrite("output.jpg", img)
# convert between BGR and RGB colour spaces
from cv2 import cvtColor, COLOR_BGR2RGB 
img_rgb = cvtColor(img, COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
import cv2
# convert from RGB to gray scale
from cv2 import COLOR_RGB2GRAY 
img = cv2.imread('img_31.jpg')
print('Before Convertion :',img[0, 0])
img_gray = cv2.cvtColor(img, COLOR_RGB2GRAY) 
cv2.imshow("Grayscale Image", img_gray)
print('After Convertion :',img_gray[0, 0])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 
# convert from RGB to gray scale in matplotlib
from cv2 import cvtColor, COLOR_BGR2RGB 
from cv2 import COLOR_RGB2GRAY
import matplotlib.pyplot as plt
img = cv2.imread('img_31.jpg') 
print('Before Convertion :',img[0, 0])
img_gray = cv2.cvtColor(img, COLOR_RGB2GRAY) 
print('After Convertion :',img_gray[0, 0])
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.title('Displaying image using Matplotlib img_rgb')
plt.show()
```
